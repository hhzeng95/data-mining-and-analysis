Be sure to document and test your function.

````python
def ndb_report(ndb_number):
    ''''Get the list of nutrients for foods.
    
    Arguments:
        NDB numbers of food
        
    Returns:
        the list of nutrients for the food
        
    '''''
    base_url = 'https://api.nal.usda.gov/ndb/V2/reports'
    params = {'api_key':'cOADnSR7bpb34RKXSasAcMsKWeWSthsFBd5nTr4O',
              'ndbno': ndb_number,  
              'type':'b',
              'format':'json'
             }
    response = requests.get(base_url, params)
    c = response.json()
    b = c['foods']
    return b[0]['food']['nutrients']
````
